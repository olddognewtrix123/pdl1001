```python
# Inspired by: https://codeocean.com/explore/capsules?query=tag:data-curation

def lipinski(smiles, verbose=False):

    moldata= []
    for elem in smiles:
        mol=Chem.MolFromSmiles(elem) 
        moldata.append(mol)
       
    baseData= np.arange(1,1)
    i=0  
    for mol in moldata:        
       
        desc_MolWt = Descriptors.MolWt(mol)
        desc_MolLogP = Descriptors.MolLogP(mol)
        desc_NumHDonors = Lipinski.NumHDonors(mol)
        desc_NumHAcceptors = Lipinski.NumHAcceptors(mol)
           
        row = np.array([desc_MolWt,
                        desc_MolLogP,
                        desc_NumHDonors,
                        desc_NumHAcceptors])   
    
        if(i==0):
            baseData=row
        else:
            baseData=np.vstack([baseData, row])
        i=i+1      
    
    columnNames=["MW","LogP","NumHDonors","NumHAcceptors"]   
    descriptors = pd.DataFrame(data=baseData,columns=columnNames)
    
    return descriptors
```

```python
df_lipinski = lipinski(df_clean_smiles.canonical_smiles)
df_lipinski
```

```python
### **Convert IC50 to pIC50**
 ### To allow **IC50** data to be more uniformly distributed, we will convert **IC50** to the negative logarithmic scale which is essentially **-log10(IC50)**.

 ### This custom function pIC50() will accept a DataFrame as input and will:
 ### * Take the IC50 values from the ``standard_value`` column and converts it from nM to M by multiplying the value by 10$^{-9}$
 ### * Take the molar value and apply -log10
 ### * Delete the ``standard_value`` column and create a new ``pIC50`` column

# https://github.com/chaninlab/estrogen-receptor-alpha-qsar/blob/master/02_ER_alpha_RO5.ipynb

import numpy as np

def pIC50(input):
    pIC50 = []

    for i in input['standard_value_norm']:
        molar = i*(10**-9) # Converts nM to M
        pIC50.append(-np.log10(molar))

    input['pIC50'] = pIC50
    x = input.drop('standard_value_norm', 1)
        
    return x
```

```python
def norm_value(input):
    norm = []

    for i in input['standard_value']:
        if i > 100000000:
          i = 100000000
        norm.append(i)

    input['standard_value_norm'] = norm
    x = input.drop('standard_value', 1)
        
    return x
```

```python
# the Mann-Whitney U Test

def mannwhitney(descriptor, verbose=False):
  # https://machinelearningmastery.com/nonparametric-statistical-significance-tests-in-python/
  from numpy.random import seed
  from numpy.random import randn
  from scipy.stats import mannwhitneyu

# seed the random number generator
  seed(1)

# actives and inactives
  selection = [descriptor, 'bioactivity_class']
  df = df_final[selection]
  intermediate = df[df['bioactivity_class'] == 'intermediate']
  intermediate = intermediate[descriptor]

  selection = [descriptor, 'bioactivity_class']
  df = df_final[selection]
  inactive = df[df['bioactivity_class'] == 'inactive']
  inactive = inactive[descriptor]

# compare samples
  stat, p = mannwhitneyu(intermediate, inactive)
  #print('Statistics=%.3f, p=%.3f' % (stat, p))

# interpret
  alpha = 0.05
  if p > alpha:
    interpretation = 'Same distribution (fail to reject H0)'
  else:
    interpretation = 'Different distribution (reject H0)'
  
  results = pd.DataFrame({'Descriptor':descriptor,
                          'Statistics':stat,
                          'p':p,
                          'alpha':alpha,
                          'Interpretation':interpretation}, index=[0])
  filename = 'mannwhitneyu_' + descriptor + '.csv'
  results.to_csv(filename)

  return results
```

```python
# Different Distribution (Reject H0):

# When you perform a Mann-Whitney U test, you start with two groups of data that you want to compare. The null hypothesis (H0) in this case is that there is no difference between the distributions of the two groups. In other words, the populations from which the two groups are drawn have the same distribution.
# If the p-value calculated from the test is below a certain significance level (often denoted as α), typically 0.05, you have evidence to reject the null hypothesis. This suggests that there is a statistically significant difference between the distributions of the two groups. Thus, you can conclude that the groups have different distributions.
# Same Distribution (Fail to Reject H0):

# On the other hand, if the p-value is greater than the significance level, you do not have enough evidence to reject the null hypothesis. In this case, you fail to reject the idea that the two groups come from the same distribution.
# This doesn't necessarily mean that the two groups are identical or that there is no difference at all between them. It just means that you haven't found strong enough evidence in your sample data to confidently claim that the distributions are different. The observed differences in your sample may have occurred due to random chance rather than being indicative of a true population difference.
# In summary, the Mann-Whitney U test helps you determine whether there is enough evidence to conclude that two groups have different distributions (rejecting the null hypothesis) or whether the differences observed in the sample data are not strong enough to make such a conclusion (failing to reject the null hypothesis).
```
